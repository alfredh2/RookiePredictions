# WRTE Dataset Analaysis

Now let's take a look at our wide receiver and tight end dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data frames
df = pd.read_csv('wrte.csv')
```

```python
df.head()
```

```python
df.columns
```

Let's first take away all the columns of professional stats that we are not trying to predict.

```python
df = df.drop(columns=['G_P', 'GS', 'Tgt', 'Ctch%',
       '1D', 'Succ%', 'Y/Tgt', 'R/G', 'Y/G', 'Fmb'])
```

We also want to make height numeric here as well.

```python
# Function to convert height from 'feet-inches' to inches
def height_to_inches(height):
    feet, inches = height.split('-')
    return int(feet) * 12 + int(inches)

# Apply the function to the Height column
df['Height'] = df['Height'].apply(height_to_inches)
df['Height'].head()
```

```python
df.info()
```

```python
corr = df.drop(columns=['Player', 'Tm', 'Conf', 'School', 'Drafted', 'Pos']).corr()
plt.figure(figsize=(20, 14))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for RBs')
plt.show()
```

Again, we have the obvious scrimmage yards and receiving yards correlations which could lead to multicollinearity, but I was more interested in seeing that weight has a moderately high correlation with college yards and other metrics, but not nearly as much with pro metrics.

Let's take a look at average receiving yards by conference.

```python
# Group by conference and average the receiving yards
conf_receiving_yards = df.groupby('Conf')['Yds_P'].mean().reset_index()

# Sort the DataFrame by passing yards for better visualization
conf_receiving_yards = conf_receiving_yards.sort_values('Yds_P', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Conf', y='Yds_P', data=conf_receiving_yards, palette='viridis')

# Add titles and labels
plt.title('Average Pro Receiving Yards by Conference')
plt.xlabel('Conference')
plt.ylabel('Average Pro Receiving Yards')

# Show the plot
plt.show()
```

Here, independent collges tend to lead, although it is closely followed by the SEC, and then other major conferences after that.

Now let's see if drafting your QB will give you more touchdowns.

```python
# Group by the 'Drafted' column and calculate the average TDs for each group
avg_tds_by_drafted = df.groupby('Drafted')['TD_P'].mean().reset_index()

# Rename the 'Drafted' column for better readability in the plot
avg_tds_by_drafted['Drafted'] = avg_tds_by_drafted['Drafted'].map({True: 'Drafted', False: 'Undrafted'})

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Drafted', y='TD_P', data=avg_tds_by_drafted, palette='viridis')

# Add titles and labels
plt.title('Average TDs of Drafted vs. Undrafted Receivers')
plt.xlabel('Draft Status')
plt.ylabel('Average TDs')

# Show the plot
plt.show()
```

Like QBs and RBs, drafted receivers will give you around twice as much TDs
