```python
import pandas as pd
```

```python
combine = pd.read_csv('combine.csv')
college = pd.read_csv('college_receiving_unclean.csv')
receiving = pd.read_csv('nfl_receiving_unclean.csv')
rushing = pd.read_csv('nfl_rushing_unclean.csv')
```

### Combine

```python
combine.head()
```

```python
combine = combine[(combine['Pos'] == 'WR') | (combine['Pos'] == 'TE')]
combine.head()
```

```python
combine.shape[0]
```

```python
len(combine['Player'].unique())
```

```python
duplicate_players = combine[combine.duplicated(subset='Player', keep=False)]
duplicate_players
```

```python
# Update the 'Player' name for "Mike Williams" from "Syracuse"
combine.loc[(combine['Player'] == 'Mike Williams') & (combine['School'] == 'Syracuse'), 'Player'] = 'Mike Williams_'
len(combine['Player'].unique())
```

```python
combine = combine.rename(columns={'Year': 'CombineYear'})
combine.head()
```

```python
combine.info()
```

### College

```python
college.head()
```

```python
college = college.rename(columns={'Year': 'CollegeYear'})
college.head()
```

```python
college.shape[0]
```

```python
len(college['Player'].unique())
```

```python
college[college['Player'] == 'Mike Williams']
```

```python
college.loc[(college['Player'] == 'Mike Williams') & (college['School'] == 'Syracuse'), 'Player'] = 'Mike Williams_'
college[college['Player'] == 'Mike Williams']
```

```python
grouped = college.groupby(['Player', 'School', 'Conf']).mean().reset_index()
grouped.shape[0]
```

```python
duplicate_players = grouped.groupby('Player').filter(lambda x: len(x) > 1)
duplicate_players
```

```python
# Sort the data by Player and CollegeYear to ensure the latest year comes last
college = college.sort_values(by=['Player', 'CollegeYear'])

# Get the latest school and conference info for each player
latest_info = college.groupby('Player').last().reset_index()

# Average the numeric columns for each player
numeric_avg = college.drop(columns=['Conf', 'School']).groupby('Player').mean().reset_index()

result = numeric_avg.drop(columns=['CollegeYear']).merge(latest_info[['Player', 'School', 'Conf']], on='Player')
result
```

```python
college = result
college.head()
```

```python
merged_df = pd.merge(college, combine, on='Player', how='inner')
merged_df.info()
```

```python
merged_df.head()
```

```python
merged_df.shape[0]
```

```python
combine.shape[0]
```

```python
len(merged_df['Player'].unique())
```

```python
merged_df[merged_df['School_x'] != merged_df['School_y']][['Player', 'School_x', 'School_y']]
```

```python
merged_df = merged_df.drop('School_x', axis=1)
merged_df.info()
```

### NFL Receiving

```python
receiving.head()
```

```python
receiving = receiving[(receiving['Pos'] == 'WR') | (receiving['Pos'] == 'TE')]
receiving.shape[0]
```

```python
receiving['Pos'].unique()
```

```python
merged3 = pd.merge(
    receiving,
    merged_df,
    left_on=['Player', 'Year'],
    right_on=['Player', 'CombineYear'],
    how='inner'
)
merged3.info()
```

### Full Dataframe Cleaning

```python
df = merged3
```

```python
df = df.drop(columns=['Y/R', 'Lng', 'Year', 'Att_R', 'Yds_R', 'Avg_R', 'TD_R', 'Bench', '3Cone', 'Shuttle'])
df.info()
```

```python
df = df.rename(columns={'Pos_x': 'Pos', 'G_x': 'G_P', 'Rec_x': 'Rec_P', 'Yds_x': 'Yds_P', 'TD_x': 'TD_P', 'G_y': 'G',
                        'Rec_y': 'Rec', 'Yds_y': 'Yds'})
df.info()
```

```python
df = df.drop(columns=['CombineYear', 'Pos_y'])
df = df.rename(columns={'TD_y': 'TD', 'School_y': 'School'})
df.info()
```

```python
df['40yd'].fillna(df['40yd'].median(), inplace=True)
df['Vertical'].fillna(df['Vertical'].median(), inplace=True)
df['Broad Jump'].fillna(df['Vertical'].median(), inplace=True)
df['Fmb'].fillna(0, inplace=True)
df['Round'].fillna(8, inplace=True)
df['Pick'].fillna(1, inplace=True)
df['OvrPick'] = (df['Round'] - 1) * 32 + df['Pick']
df = df.drop(columns=['Pick', 'Round'])
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
# Save the DataFrame to a CSV file
df.to_csv('wrte.csv', index=False)
```
