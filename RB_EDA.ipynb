# RB Dataset Analaysis

Now let's take a look at our running back dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data frames
df = pd.read_csv('rb.csv')
```

```python
df.head()
```

```python
df.columns
```

Let's first take away all the columns of professional stats that we are not trying to predict.

```python
df = df.drop(columns=['Tgt', 'Ctch%', '1D',
       'Succ%', 'Y/Tgt', 'R/G', 'Y/G', 'G_P', 'GS', 
       '1D_P', 'Succ%_R', 'Y/A', 'Y/G_R'])
```

We also want to make height numeric here as well.

```python
# Function to convert height from 'feet-inches' to inches
def height_to_inches(height):
    feet, inches = height.split('-')
    return int(feet) * 12 + int(inches)

# Apply the function to the Height column
df['Height'] = df['Height'].apply(height_to_inches)
df['Height'].head()
```

```python
df.info()
```

```python
corr = df.drop(columns=['Player', 'Tm', 'Conf', 'School', 'Drafted']).corr()
plt.figure(figsize=(20, 14))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for RBs')
plt.show()
```

There is obviously high correlation betweeen the scrimmage metrics and rushing metrics for college, so there will definitely be some multicollienarity issues there. Other than that, nothing stands out too much besides overall pick medium high correlations with certain metrics, which makes sense.

Let's take a look at average rushing yards by conference.

```python
# Group by conference and average the rushing yards
conf_rushing_yards = df.groupby('Conf')['Yds_R_P'].mean().reset_index()

# Sort the DataFrame by passing yards for better visualization
conf_rushing_yards = conf_rushing_yards.sort_values('Yds_R_P', ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Conf', y='Yds_R_P', data=conf_rushing_yards, palette='viridis')

# Add titles and labels
plt.title('Average Pro Rushing Yards by Conference')
plt.xlabel('Conference')
plt.ylabel('Average Pro Rushing Yards')

# Show the plot
plt.show()
```

This is quite a change from the QB dataset, where the highest passing yards were all in the most major conferences. Here, the lesser known conferences have higher rushing yards. 

Now let's see if drafting your QB will give you more rushing touchdowns.

```python
# Group by the 'Drafted' column and calculate the average TDs for each group
avg_tds_by_drafted = df.groupby('Drafted')['TD_R_P'].mean().reset_index()

# Rename the 'Drafted' column for better readability in the plot
avg_tds_by_drafted['Drafted'] = avg_tds_by_drafted['Drafted'].map({True: 'Drafted', False: 'Undrafted'})

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Drafted', y='TD_R_P', data=avg_tds_by_drafted, palette='viridis')

# Add titles and labels
plt.title('Average TDs of Drafted vs. Undrafted RBs')
plt.xlabel('Draft Status')
plt.ylabel('Average TDs')

# Show the plot
plt.show()
```

This is similar to the drafted vs undrafted QBR rankings. Drafted average TDs is nearly double, but you might expect it to be even higher.
